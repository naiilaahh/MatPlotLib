# Data Visualization witg MatPlot Lib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Line Plots
- timw series data
- trends and patterns in data
- observe linear relationships

```python
# Creating plot data
xpoint = np.array([1, 8])
ypoints = np.array([3, 10])
```

```python
# State based approach
plt.plot(xpoint, ypoints)
plt.title('First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show() # shows the plot graph even in a .py file
```

```python
# object oriented approach

fig, ax = plt.subplots()
ax.plot(xpoint, ypoints)
plt.show()
```
