# Data Visualization witg MatPlot Lib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Line Plots
- timw series data
- trends and patterns in data
- observe linear relationships

```python
# Creating plot data
xpoint = np.array([1, 8])
ypoints = np.array([3, 10])
```

```python
# State based approach
plt.plot(xpoint, ypoints)
plt.title('First Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show() # shows the plot graph even in a .py file
```

```python
# object oriented approach

fig, ax = plt.subplots()
ax.plot(xpoint, ypoints)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10,4,200,45,67,56,98,100])
```

```python
plt.plot(ypoints, marker='s', markerfacecolor='#51074a', linestyle='--' ,color='#8f83d8')
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])
```

```python
plt.plot(x, y, marker='D', markerfacecolor='#ffb6c1', linestyle=':', color='#191970')
plt.title('Title: Sports Watched Data', color='#FF9999')
plt.ylabel('Calories Burned', color='#9ACD32')
plt.xlabel('Average Pulse Rate', color='#6F00FF')
plt.show()
```

```python
years=[1970,1980,1985,1990,1995,2000,2005,2010,2015,2020,2025,2030,2035]
pops=[3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3,7.3,7.3,7.4,7.5]
```

```python
plt.plot(years, pops, marker='P', linestyle=':', color='#F2A2BD')
plt.title('World Population Graph', color='Blue')
plt.xlabel('Population Growth by Year', color='green')
plt.ylabel('Population in Billions', color='red')
plt.show()
```

### Font Styles
- fontsize
- fontweight
- font family

```python
plt.plot(x, y, marker='D', markerfacecolor='#ffb6c1', linestyle=':', color='#191970')
plt.title('Title: Sports Watched Data', color='#FF9999', fontweight='bold', fontsize=20)
plt.ylabel('Calories Burned', color='#9ACD32', fontfamily='monospace')
plt.xlabel('Average Pulse Rate', color='#6F00FF')
plt.show()
```

```python
# orecreated font class
label_font = {
    'family': 'sans serif',
    'color': 'pink',
    'weight': 'normal',
    'size':20
}

plt.plot(x, y, marker='D', markerfacecolor='#ffb6c1', linestyle=':', color='#191970')
plt.title('Title: Sports Watched Data', color='#FF9999')
plt.ylabel('Calories Burned', color='#9ACD32')
plt.xlabel('Average Pulse Rate', fontdict=label_font)
plt.show()
```

```python
plt.plot(ypoints, marker='D', linestyle='--')
plt.grid(color='red', linestyle='-', alpha=0.25)  # alpha = opacity of the color
plt.show()
```

```python
# coloring line with kwargs - key word argument
plt.plot(x, y, marker='D', linestyle=':', color='blue')

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 130, 5))
plt.ylabel('Calorie Burnage')

plt.grid(color='red')
plt.gca().set_axisbelow(True) # sets gridlines behihnd plot elements

plt.gca().set_facecolor('#FFD8F0')

plt.show()
```

## Controls the size & properties of the figure (fig/canvas)

```python
fig = plt.figure(figsize=(6,4), dpi=150) # figsize to adjust the size and shape of your figure

plt.plot(x, y, marker='D', linestyle=':', color='blue')

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 130, 5))
plt.ylabel('Calorie Burnage')

plt.savefig('sportswatchdata') # saves graph to file

plt.show()
```

### multi plot lines

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
job_rate = [0.5, 0.6, 0.8, 0.9, 1.0, 1.2, 1.5, 1.7, 2.0, 2.2, 2.5, 2.7, 2.9]
```

```python
plt.figure(figsize=(10,8))

plt.plot(years, pops, color='#FF6FFF', linestyle=':', marker='d', label='population')
plt.plot(years, death, color='#FC0FC0', linestyle='--', marker='s', label='deaths')
plt.plot(years, job_rate, color='#E0218A', linestyle='dashdot', marker='o', label='job_rate')
plt.grid(linestyle='--')

plt.title('Population Growth')
plt.xlabel('Year')
plt.ylabel('Population Growth by Year (Billions)')

plt.legend() # creates key. must label plt.plot using label=

plt.show()
```
