```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print('Missing Values: ')
employee.isnull().sum()
```

```python
# show duplicate records 
print('Duplicates: ')
employee.duplicated().sum()
```

```python
# Filling in missing values 
employee['Name'] = employee['Name'].fillna('Unknown')

print('Age Mean', employee['Age'].mean())
print('Age Median', employee['Age'].median())
print('Age Mode', employee['Age'].mode())
```

```python
age_median = employee['Age'].median()
employee['Age'] = employee['Age'].fillna(age_median)
```

```python
print(' Weight Mean', employee['Weight'].mean())
print('Weight Med', employee['Weight'].median())
print('Weight Most', employee['Weight'].mode())
```

```python
weight_median = employee['Weight'].median()
employee['Weight'] = employee['Weight'].fillna(weight_median)
```

```python
salary_median = employee['Salary'].median()
employee['Salary'] = employee['Salary'].fillna(salary_median)
```

```python
employee
```

```python
employee.isnull().sum()
```

```python
employee = employee.sort_values(by='Age', ascending=True)

plt.figure(figsize=(10,6))
plt.plot(employee['Age'], employee['Salary'], marker='o')
plt.xlabel('Age')
plt.ylabel('Salary (In Millions)')
plt.title('LineGraph of Salary by Age')
plt.grid(linestyle=':')


plt.show()
```
