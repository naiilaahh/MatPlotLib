# RLAB 386.5.1 Plotting Line Styles

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
```

```python
# Solid linestyle
plt.plot(xpoints, ypoints, linestyle='-')
plt.title('Lab Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Dashed linestyle
plt.plot(xpoints, ypoints, linestyle='--')
plt.title('Lab Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Dashdot linestyle
plt.plot(xpoints, ypoints, linestyle='dashdot')
plt.title('Lab Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Dotted linestyle
plt.plot(xpoints, ypoints, linestyle=':')
plt.title('Lab Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(xpoints, ypoints, linestyle=':', color='#C54B8C', marker='D')
plt.show()
```
