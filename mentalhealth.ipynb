```python
df = df.sort_values(by='daily_social_media_hours', ascending=True)
plt.figure(figsize=(10,6))
plt.plot(df['daily_social_media_hours'], df['academic_performance'], marker='o')
plt.xlabel('Daily Social Media Time')
plt.ylabel('Academic Performance')
plt.title('LineGraph of Academic performance by SM time')

plt.grid(linestyle=':')

plt.show()
```

```python
df_agg = df.groupby('age', as_index=False)[['daily_social_media_hours', 'academic_performance', 'sleep_hours', 'stress_level', 'physical_activity']].median()
df_agg
```

```python
plt.plot(df_agg['age'], df_agg['daily_social_media_hours'], color='#FF9999', label='Social Media Hours')
plt.plot(df_agg['age'], df_agg['sleep_hours'], color='#FF6FFF', label='Sleep Hours')
plt.plot(df_agg['age'], df_agg['academic_performance'], color='#E0218A', label='Academic Performance')

plt.title('Teen Age vs Hours of Academic Performance, Sleep, Social Media')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.grid(linestyle='--')
plt.legend(loc=7)
plt.show()
```
